## 2. Penggabungan Data Menggunakan Inner Join, Left Join, Right Join, dan Outer Join (Studi Kasus: Breast Cancer Dataset)

```python
import pandas as pd
df1breastquad_odd = pd.read_csv("/content/df1breastquad_odd.csv", sep=';')
df2breastquad_even = pd.read_csv("/content/df2breastquad_even.csv", sep=';')
df1tumor_size_odd = pd.read_csv("/content/df1tumor_size_odd.csv", sep=';')
df2tumor_size_even = pd.read_csv("/content/df2tumor_size_even.csv", sep=';')

print(df1breastquad_odd)
print(df2breastquad_even)
print(df1tumor_size_odd)
print(df2tumor_size_even)
```

```python
breast = pd.concat([df1breastquad_odd, df2breastquad_even]).sort_values('PatientID')
breast
```

```python
tumor = pd.concat([df1tumor_size_odd, df2tumor_size_even]).sort_values('PatientID')
tumor
```

```python
inner_join = pd.merge(tumor, breast, on = "PatientID", how = 'inner')
outer_join = pd.merge(tumor, breast, on = "PatientID", how = 'outer')
left_join = pd.merge(tumor, breast, on = "PatientID", how = 'left')
right_join = pd.merge(tumor, breast, on = "PatientID", how = 'right')

print(inner_join)
print(outer_join)
print(left_join)
print(right_join)
```

Berapa banyak record pasien yang memiliki nilai atribut ukuran tumor dan nilai

```python
total = inner_join.shape[0]
print('Jumlah banyak record pasien yang memiliki nilai atribut ukuran tumor dan nilai kuadran kanker payudara secara lengkap keduanya adalah', total)
```

# 4. Deteksi Outlier: InterQuartile Range dan BoxPlot

```python
import pandas as pd
atm_transaction = pd.read_csv("/content/ATM-Transaction.csv", sep=';')
atm_transaction
```

```python
print("Statistik Deskriptif:")
atm_transaction.describe()
```

```python
bank_withdrawals = atm_transaction.groupby('ATM Name')['Total amount Withdrawn (Rupiah)'].sum()
teramai = bank_withdrawals.idxmax()
tersepi = bank_withdrawals.idxmin()
print("Bank teramai untuk penarikan uang", teramai)
print("Bank tersepi untuk penarikan uang", tersepi)
```

```python
atm_transaction['Transaction Date'] = pd.to_datetime(atm_transaction['Transaction Date'])
day_of_week_counts = atm_transaction['Transaction Date'].dt.day_name().value_counts()
sering_dikunjungi = day_of_week_counts.idxmax()
jarang_dikunjungi = day_of_week_counts.idxmin()
print("Hari paling sering dikunjungi untuk penarikan uang adalah", sering_dikunjungi)
print("Hari paling jarang dikunjungi untuk penarikan uang adalah", jarang_dikunjungi)
```

```python
Q1 = atm_transaction['Total amount Withdrawn (Rupiah)'].quantile(0.25)
Q3 = atm_transaction['Total amount Withdrawn (Rupiah)'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = atm_transaction[(atm_transaction['Total amount Withdrawn (Rupiah)'] < lower_bound) | (atm_transaction['Total amount Withdrawn (Rupiah)'] > upper_bound)]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.boxplot(atm_transaction['Total amount Withdrawn (Rupiah)'])
plt.title('Box Plot - Total amount Withdrawn (Rupiah)')
plt.show()

# Nama ATM dan No Of Withdrawals yang terdeteksi sebagai outlier
print("ATM dan No Of Withdrawals yang terdeteksi sebagai outlier:")
print(outliers[['ATM Name', 'No Of Withdrawals']])
```
